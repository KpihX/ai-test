```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contains {x.shape[0]} examples and {x.shape[1]} variables')
print(f'There are {np.unique(y).size} classes')
print(names)
```

## Visualisation

```python
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel('longueur s√©pal')
plt.ylabel('largeur s√©pal')
```

### 3D visualisation

```python
%matplotlib
from mpl_toolkits.mplot3d import Axes3D

ax = plt.axes(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
```

### Hist visualisation

```python
plt.hist(x[:, 0], bins=20)
plt.hist(x[:, 1], bins=20)
```

```python
plt.hist2d(x[:, 0], x[:, 1], cmap='Blues')
plt.xlabel('longueur s√©pal')
plt.ylabel('largeur s√©pal')
plt.colorbar()
```

### Imshow representation

```python
plt.imshow(np.corrcoef(x.T), cmap='Blues')
plt.colorbar()
```

### Exo: visualisation vars

```python
from scipy.special import comb
def graphics(x, y):
    n = x.shape[1]
    n_subplots = comb(n, 2, exact=True)

    k = 0
    plt.figure(figsize=(10, 40))
    for i in range(1, n+1):
        for j in range(i+1, n+1):
            k += 1
            plt.subplot(n_subplots, 1, k)
            plt.scatter(x[:, i-1], x[:, j-1], c=y, alpha=0.5)
            plt.xlabel(f'x_{i-1}')
            plt.ylabel(f'x_{j-1}')
            plt.title(f'x_{i-1} & y_{j-1}')

graphics(x, y)

```
